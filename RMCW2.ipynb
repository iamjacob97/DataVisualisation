Importing all the necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
%matplotlib inline
```

Converting data into a pandas dataframe

```python
df = pd.read_csv("Results_21Mar2022.csv")
df
```

Descriptive stats

```python
print(f"Dataset shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
print("\nData types:")
print(df.dtypes)
print("\nSummary statistics:")
print(df.describe())
print("\nMissing values:")
print(df.isnull().sum())
```

```python
print("Unique values in diet_group:", df['diet_group'].unique())
print("Unique values in sex:", df['sex'].unique())
print("Unique values in age_group:", df['age_group'].unique())
print("Number of unique groupings:", df['grouping'].nunique())
print("Sample of groupings:", df['grouping'].unique()[:10])
```

From the data above we can see that the data shows 1000 runs of the similar experiments studied on 72 composite groups of different combinations of diet-type(vegan, meat, fish etc), sex(male, female), and age ranges ranging from 20 - 79. Let's start by examining the relationship between diet groups and environmental impacts.

```python
metrics = ['mean_ghgs', 'mean_land', 'mean_watscar', 'mean_eut', 
           'mean_ghgs_ch4', 'mean_ghgs_n2o', 'mean_bio', 'mean_watuse', 'mean_acid']
diet_impacts = df.groupby('diet_group')[metrics].mean().reset_index()

print(diet_impacts)

normalized_impacts = diet_impacts.copy()
for column in normalized_impacts.columns[1:]:
    normalized_impacts[column] = normalized_impacts[column] / normalized_impacts[column].max()

print("\nNormalized environmental impacts by diet group:")
print(normalized_impacts)
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(normalized_impacts.set_index('diet_group'), cmap='viridis', 
            annot=True, fmt='.2f', linewidths=.5)

plt.title('Normalized Environmental Impacts by Diet Type', fontsize=16)
plt.ylabel('Diet Type', fontsize=12)
plt.tight_layout()
plt.show()
```

Key observations:

- Meat100 consistently shows the highest impact (all 1.00) across all environmental metrics.  
- Vegan diet consistently shows the lowest impact across nearly all metrics (especially in methane emissions at just 0.06).  
- Water scarcity (mean_watscar) is surprisingly high across all diet types (even vegan at 0.68).  
- Water usage (mean_watuse) shows substantial variation with meat and meat100 having the highest impact.  
- Land use (mean_land) shows dramatic differences between diet types, with meat100 using over 5 times more land than vegan diets.  

Now, let's examine how sex influences these environmental impacts:

```python
sex_impacts = df.groupby("sex")[metrics].mean().reset_index()
print(sex_impacts)

normalized_impacts = sex_impacts.copy()
for column in normalized_impacts.columns[1:]:
    normalized_impacts[column] = normalized_impacts[column] / normalized_impacts[column].max()

print("\nNormalized environmental impacts by sex group:")
print(normalized_impacts)
```

```python
normalized_impacts.set_index('sex').T.plot(kind='bar', figsize=(12, 6))
plt.title('Normalized Environmental Impacts by Sex')
plt.ylabel('Normalized Impact (0-1)')
plt.xlabel('Environmental Metric')
plt.legend(title='Sex')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Males always dominate but not by a significant margin

Age Group Influence

```python
age_impacts = df.groupby("age_group")[metrics].mean().reset_index()
print(age_impacts)

normalized_impacts = age_impacts.copy()
for column in normalized_impacts.columns[1:]:
    normalized_impacts[column] = normalized_impacts[column] / normalized_impacts[column].max()

print("\nNormalized environmental impacts by age group:")
print(normalized_impacts)
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(normalized_impacts.set_index('age_group'), cmap='viridis', 
            annot=True, fmt='.2f', linewidths=.5)

plt.title('Normalized Environmental Impacts by Age Group', fontsize=16)
plt.ylabel('Age Group', fontsize=12)
plt.tight_layout()
plt.show()
```

From the above observations it is clear that the higher values for environmental impact factors tend to be towards meat100, males of 70 - 79 age groups. Let's aggregate the data of the 1000 different runs into 72 rows averaging the metrics across runs, as we’ve done before.

```python
grouped = df.groupby('grouping').agg({ **{metric: 'mean' for metric in metrics}, 
    'diet_group': 'first', 'sex': 'first', 'age_group': 'first'}).reset_index()

print(grouped.head())
print("Rows:", grouped.shape[0])
```

We will normalise the data like before using a scaler from scikit-learn


```python
from sklearn.preprocessing import MinMaxScaler

normalized_grouped = grouped.copy()

scaler = MinMaxScaler()
normalized_grouped[metrics] = scaler.fit_transform(grouped[metrics])

print(normalized_grouped.head())
```

We know that we have observed the highest variance in the impact factors with respect to the diet type. So I will colour code them for maximum visual clarity, sex and age group will be encoded by line type and opacity respectively.

```python
diet_colors = {'vegan': 'green', 'veggie': 'lightgreen', 'fish': 'blue', 
               'meat': 'orange', 'meat50': 'yellow', 'meat100': 'red'}
sex_styles = {'female': '--', 'male': '-'}

age_groups = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
age_alphas = [0.33, 0.42, 0.57, 0.73, 0.86, 1.0]
age_alpha_map = dict(zip(age_groups, age_alphas))

normalized_grouped['alpha'] = normalized_grouped['age_group'].map(age_alpha_map)
```

```python
from matplotlib.lines import Line2D
import mplcursors
%matplotlib widget

metric_labels = {
    'mean_ghgs': 'Greenhouse Gases',
    'mean_land': 'Land Use',
    'mean_watscar': 'Water Scarcity',
    'mean_eut': 'Eutrophication',
    'mean_ghgs_ch4': 'GHGs (CH4)',
    'mean_ghgs_n2o': 'GHGs (N2O)',
    'mean_bio': 'Biodiversity Impact',
    'mean_watuse': 'Water Use',
    'mean_acid': 'Acidification'
}

readable_metrics = [metric_labels.get(m, m) for m in metrics]
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(13, 7))

    plotted_lines = []
    line_to_data_map = {}

    for _, row in normalized_grouped.iterrows():
        current_alpha = age_alpha_map.get(row['age_group'], 0.7)
        line, = plt.plot(range(len(metrics)), row[metrics], color=diet_colors[row['diet_group']], linestyle=sex_styles[row['sex']],
                         alpha=current_alpha, linewidth=1.135, marker='o', markersize=3, label='_nolegend_')

        plotted_lines.append(line)
        line_to_data_map[line] = (row['diet_group'], row['sex'], row['age_group'])
    
    plt.xticks(range(len(metrics)), readable_metrics, rotation=45, ha='right', fontsize=11)    
    plt.ylabel('Normalized Impact (0-1)', fontsize=13)
    plt.title('Environmental Impacts Across Diet, Sex, and Age Groups (Hover for details)', fontsize=17, pad=21)
    
    diet_legend = [Line2D([0], [0], color=color, lw=2, label=diet) for diet, color in diet_colors.items()]
    sex_legend = [Line2D([0], [0], color='lightgray', linestyle=style, lw=2, label=sex) for sex, style in sex_styles.items()]
    age_legend = [plt.Rectangle((0,0), 1, 1, fc="lightgray", alpha=0.7, ec='gray', label='Age (Opacity)')] 
    all_handles = diet_legend + sex_legend + age_legend

    legend = plt.legend(handles=all_handles, title='Legend', loc='center left', bbox_to_anchor=(1.0, 0.5), 
               fontsize=9, title_fontsize=11, facecolor='dimgray', edgecolor='darkgray') 
    plt.setp(legend.get_texts(), color='white') 
    plt.setp(legend.get_title(), color='white') 

    plt.tight_layout(rect=[0, 0, 0.89, 1]) 

    plt.grid(True, linestyle='--', color='gray', alpha=0.3) 
    
    cursor = mplcursors.cursor(plotted_lines, hover=True)

    @cursor.connect("add")
    def on_add(sel):
        diet, sex, age = line_to_data_map[sel.artist]
        sel.annotation.set(text=f'Diet: {diet}\nSex: {sex}\nAge: {age}')
        sel.annotation.get_bbox_patch().set(fc="whitesmoke", alpha=0.9) 
        sel.annotation.set_color('black') 
    
plt.show() 
```
